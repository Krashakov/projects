### Обработка пропусков

```python
data.isna().sum()
```

Изучим строки с отсутствующим годом выпуска игры.

```python
data.loc[data['year_of_release'].isna()].tail(10)
```

Видим, что в названии некоторых игр (как правило, спортивных симуляторов) указан год. Такая маркировка связана с тем, что игры этого жанра (симуляторы) выходят ежегодно. Чтобы различать игры с одним названием, и указывают дату выхода.  

Но необходимо учитывать, что создатели игр часто выпускают игры с "опережением", т.е. в 2009 году выпускают, допустим, игру FIFA 2010. Это делается в рекламных целях (для создания ощущения новизны игры), т.к. часто релиз приходится на конец года, и игра может "устареть" через несколько месяцев.

С помощью лемматизации "достанем" год из названия игры и пропишем его в столбце "Год выпуска" с учетом трюка издателей (вычтем из года в названии "единицу").

```python
m = Mystem() 

def year_lem(row):
    name = row['name']
    lemmas = m.lemmatize(name)
    return Counter(lemmas)   

data['year_lem'] = data.apply(year_lem, axis=1)  
data.head()
```

```python
def year_group(year_lem):
    for i in range(1980, 2017):
        j=str(i)
        if j in year_lem:    
            return i-1
    
data['year_group'] = data['year_lem'].apply(year_group)
data.head()
```

```python
data.loc[~ data['year_group'].isna()].head(10)
```

```python
data.info()        
        
```

```python
data['year_of_release'] = data['year_of_release'].fillna(data['year_group']) 
```

```python
data.isna().sum()
```

Видим, что удалось заполнить 17 строк. Маловато, конечно. Похоже, быстрее это было сделать вручную) Зато год заполнен в этих случаях с минимальной погрешностью.

Оставшиеся пропуски по году выпуска удалим.   


```python
data = data.loc[~ data['year_of_release'].isna()].reset_index(drop=True)
```

```python
data.isna().sum()
```

Теперь изучим пропуски в столбцах с оценками и рейтингом.

```python
data.query('1980 < year_of_release < 2001').head(10)
```

```python
data.query('not critic_score.isna() and 1980 < year_of_release < 2001').count()
```

Резкий рост игр с оценками наблюдается в районе 2000 года. Это означает, что сервис, данные с которого приводятся в таблице, возник где-то в это время. Возможно, это сайт Metacritic.com, где собраны оценки критиков и пользователей для многих игр, фильмов и т.д.



В столбце с оценками игроков видим значение "tbd" (to be determined). Этот термин часто встречается, когда голосование (выставление оценок) еще закрыто для пользователей ввиду недавнего выхода игры.  

```python
data.query('user_score == "tbd"').head(10)
```

Видим, что статус tbd также встречается у непопулярных игр (суммарные продажи составляют около 10 тыс.). Скорее всего, эти игры просто не набрали необходимое для выставления оценки количество голосов пользователей.

Исключим игры со значением tbd.

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan)
```

```python
data.isna().sum()
```

```python
data_scores = data.loc[data['user_score'] != 'tbd'].reset_index(drop=True)
data_scores.head(10)
```

Приведем столбец с оценками игроков к типу float (т.к. у этих оценок есть дробная часть).

```python
data_scores['user_score'] = data_scores['user_score'].astype('float')
data_scores.info()
```

Умножим оценки юзеров на 10, чтобы привести их к одному виду с оценками критиков (100-балльная система).

```python
data_scores['user_score'] = data_scores['user_score'] * 10
data_scores.head(10)
```

Посчитаем корреляцию между двумя типами оценок.

```python
data_scores['critic_score'].corr(data_scores['user_score'])
```

```python
sns.set_style('darkgrid')
```

```python
data_scores.plot(x='critic_score', y='user_score', title='Оценки пользователей и оценки критиков', style='o',
    grid=True, figsize=(9, 6), alpha=0.2);
```

```python
data_scores[['user_score', 'critic_score']].plot.hist(bins=25, range=(0, 100), figsize=(8, 5), alpha=0.7)
plt.title('Оценки критиков и пользователей')
plt.xlabel('Оценки критиков и пользователей')
plt.ylabel('Количество игр')
plt.show;
```

Видим, что связь между оценками пользователей и оценками критиков умеренная. В целом, оценки юзеров ниже оценок критиков на средне-низких значениях, но выше - на высоких (зато у критиков есть очень высокие оценки - около 100). Основная масса значений находится в районе 65-85. Нередки случаи, когда мнения критиков и игроков расходятся (при высоких оценках профессионалов мы видим довольно низкие оценки юзеров). 

Таким образом, связь между двумя типами оценок недостаточно сильна, чтобы мы могли по одному виду заполнить другой. Т.е. мы вынуждены оставить пропуски по оценкам, чтобы не исказить реальную картину, а для расчетов по оценкам довольствоваться теми играми, для которых даны оба вида.

## Портрет пользователя каждого региона

Чтобы изучить предпочтения игроков из разных регионов (Америка, Европа и Япония), определим игры, популярные в каждом регионе. Для этого рассчитаем "коэффициент популярности", разделив продажи в регионе на суммарные. 

### Самые популярные платформы

```python
data_region = data.drop(['critic_score', 'user_score'], axis=1)
```

```python
data_region['na_ratio'] = data_region['na_sales'] / data_region['total_sales']
data_region['eu_ratio'] = data_region['eu_sales'] / data_region['total_sales']
data_region['jp_ratio'] = data_region['jp_sales'] / data_region['total_sales']
data_region.head()
```

Получили таблицу с "коэффициентами популярности" (КП) для каждого региона. Теперь категоризируем игры по следующему принципу: если КП в каком-то регионе > 0.5, тогда отнесем игру к этому региону. Если же таких коэффициентов для игры нет, отнесем ее к "Другому" региону.   

```python
def region(row):
    na_ratio = row['na_ratio']
    eu_ratio = row['eu_ratio']
    jp_ratio = row['jp_ratio']
    if na_ratio > 0.5:
        return 'Америка'
    if eu_ratio > 0.5:
        return 'Европа'
    if jp_ratio > 0.5:
        return 'Япония'
    return 'Другой'

data_region['region'] = data_region.apply(region, axis=1)
data_region.head()
```

```python
data_region.groupby('region')['name'].count()
```

Получили, что больше всего игр в американском регионе. Выглядит довольно логично, учитывая, что большинство студий-разработчиков игр находится в США (и в Канаде). Многие студии ориентируются именно на внутренний рынок, выпуская продукт, интересный в первую очередь американской аудитории. Также следует сказать, что одной из причин хороших продаж в Северной Америке является высокая платежеспособность игроков (цена топовых игр может доходить до 80$).

## Проверка гипотез

###  Пользовательские рейтинги Xbox One и PC

Проверим гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и РС одинаковые.

Нулевая гипотеза (Н0): средние пользовательские рейтинги платформ Xbox One и PC не различаются.

Альтернативная гипотеза (Н1): средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
data_scores.head()
```

```python
data_scores_XOne = data_scores.loc[(data_scores['platform'] == 'XOne') & (~ data_scores['user_score'].isna())]
data_scores_PC = data_scores.loc[(data_scores['platform'] == 'PC')  & (~ data_scores['user_score'].isna())]
```

Проверим распределения пользовательских оценок XOne и PC на нормальность.

```python
plt.figure(figsize=(8, 5))
sns.distplot(data_scores_XOne['user_score'], bins=20)
sns.distplot(data_scores_PC['user_score'], bins=20)
plt.title('Оценки пользователей (Xbox One и PC)')
plt.xlabel('Оценки пользователей')
plt.show;
```

Оценки распределены нормально, поэтому можно применять t-критерий Стьюдента для независимых выборок.
Сначала подсчитаем дисперсию и стандартное отклонение. У нас в распоряжении не вся генеральная совокупность, а выборка из нее, поэтому в методе var укажем параметр ddof=1.


```python
np.var(data_scores_XOne['user_score'], ddof=1)
```

```python
np.std(data_scores_XOne['user_score'], ddof=1)
```

```python
np.var(data_scores_PC['user_score'], ddof=1)
```

```python
np.std(data_scores_PC['user_score'], ddof=1)
```

```python
#зададим критический уровень статистической значимости, равный 1% (высокий уровень значимости)
alpha = 0.01
results = st.ttest_ind(data_scores_XOne['user_score'], data_scores_PC['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Для проверки применим непараметрический критерий Манна-Уитни.

```python
alpha = 0.01
results = st.mannwhitneyu(data_scores_XOne['user_score'], data_scores_PC['user_score'], alternative='two-sided')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Мы можем отвергнуть нулевую и принять альтернативную гипотезу, т.е. средние пользовательские оценки платформ Xbox One и PC различаются.
